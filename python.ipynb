```python
# plot the separate and pooled models
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(10, 8))
x = np.linspace(1, 7, 7)

# separate
ax = axes[0]
lines = ax.plot(
    x,
    [0.5/y for y in x],
    color='C0',
    linewidth=0.5,
    label='avgerage learning time'
)
# highlight the last line
lines[-1].set_linewidth(2)
lines[-1].set_color('b')
ax.legend()
ax.set_ylim((0, ax.set_ylim()[1]))  # set y base to zero
# ax.set_yticks(())
ax.set_xlabel('days')
ax.set_ylabel('learning time(h)')
ax.set_title('app version=1.0.0')

# separate
ax = axes[1]
lines = ax.plot(
    x,
    [0.4/y for y in x],
    color='C0',
    linewidth=0.5,
    label='avgerage learning time'
)
# highlight the last line
lines[-1].set_linewidth(2)
lines[-1].set_color('r')
ax.legend()
ax.set_ylim((0, ax.set_ylim()[1]))  # set y base to zero
# ax.set_yticks(())
ax.set_xlabel('days')
ax.set_ylabel('learning time(h)')
ax.set_title('app version=1.1.0')
```
